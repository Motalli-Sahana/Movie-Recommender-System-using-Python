```python
import numpy as np
import pandas as pd
```

```python
column_names=['user_id','item_id','rating','timestamp']
df=pd.read_csv('/content/sample_data/u.data',sep='\t',names=column_names)
```

```python
df.head()
```

```python
movie_titles=pd.read_csv('/content/sample_data/Movie_Id_Titles.txt')
movie_titles.head()
```

```python
df=pd.merge(df,movie_titles,on='item_id')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('white')
%matplotlib inline
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings=pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['number of ratings']=pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
plt.figure(figsize=(10,4))
ratings['number of ratings'].hist(bins=70)
```

```python
plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating',y='number of ratings',data=ratings,alpha=0.5)
```

```python
moviemat=df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
ratings.sort_values('number of ratings',ascending=False).head(10)
```

```python
ratings.head()
```

```python
starwars_user_ratings=moviemat['Star Wars (1977)']
toystory_user_ratings=moviemat['Toy Story (1995)']
starwars_user_ratings.head()
```

```python
similar_to_starwars=moviemat.corrwith(starwars_user_ratings)
similar_to_toystory=moviemat.corrwith(toystory_user_ratings)
```

```python
corr_starwars=pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars=corr_starwars.join(ratings['number of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['number of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_toystory=pd.DataFrame(similar_to_toystory,columns=['Correlation'])
corr_toystory.dropna(inplace=True)
corr_toystory.head()
```

```python
corr_toystory=corr_toystory.join(ratings['number of ratings'])
corr_toystory.head()
```

```python
corr_toystory[corr_toystory['number of ratings']>100].sort_values('Correlation',ascending=False).head()
```
